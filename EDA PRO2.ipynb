```python
print("ðŸ”¹ Cleaning Column Names...")
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
print(f"âœ… Cleaned Columns: {list(df.columns)}\n")
```

```python
# === Step 5: Analyzing Numeric Columns ===
numeric_columns = df.select_dtypes(include=['number']).columns
print(f"ðŸ”¹ Numeric Columns: {list(numeric_columns)}\n")
```

```python
# Plot histograms for numeric features
for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=20, kde=True)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Distribution of {col}")
    plt.show()
    
```

```python
# === Step 6: Analyzing Categorical Columns ===
categorical_columns = df.select_dtypes(include=['object']).columns
print(f"ðŸ”¹ Categorical Columns: {list(categorical_columns)}\n")
```

```python
# Display unique values for categorical columns
for col in categorical_columns:
    print(f"ðŸ”¹ Unique values in '{col}': {df[col].nunique()}")
    print(df[col].value_counts().head(), "\n")
```

```python
if len(numeric_columns) > 1:
    plt.figure(figsize=(10, 6))
    sns.heatmap(df[numeric_columns].corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("ðŸ”¹ Correlation Heatmap")
    plt.show()
```
